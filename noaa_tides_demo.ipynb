# Loop In Class Activity with readline()
Loop through first 55 lines of file

## For Loop

```python
# read data
filepath='data/noaa_tides.txt'
file_obj=open(filepath,'r')

# set cut off variable to stop counting after 55
cutoff=55
# create loop
for line in range(0,cutoff):
    print(file_obj.readline())
```

## While loop

```python
# read data
filepath='data/noaa_tides.txt'
file_obj=open(filepath,'r')

# set variables for boolean condition
cutoff=55
counter=0
# create loop
while counter < cutoff:
    print(file_obj.readline())
    counter +=1
```
